```python
from itertools import chain
```

```python
def show_recom(cust_id,number_of_rec):
    products_overall=df["ProductsID"].unique() # collecting al the productid
    procust_bought=df[df["CustomerId"]==cust_id]["ProductsID"].values #finding the number of products purchased by the customer
    prod_used_for_predict=[i for i in products_overall if i not in procust_bought] # not purchased by the customer
    predict_recom=[(prod,model.predict(cust_id,prod).est)for prod in prod_used_for_predict] #products and rating of the product(.est)

    output_recom= [df[df["ProductsID"]==i_d]["Products"].values[0] for i_d,a in predict_recom[:number_of_rec]]
    out_list =list(chain.from_iterable(output_recom))
    out_list=list(set(out_list))
    return out_list
```
